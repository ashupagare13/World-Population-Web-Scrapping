```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = 'https://www.worldometers.info/world-population/population-by-country/'
```

```python
response = requests.get(url)
```

```python
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
print(soup)
```

```python
rows = soup.find('table', {'id':'example2'}).find('tbody').find_all('tr')
```

```python
print(len(rows))
```

```python
population_list = []


for r in rows:
    dict = {}
    
    dict['country'] = r.find_all('td')[1].text
    dict['population 2020'] = r.find_all('td')[2].text.replace(',','')
    dict['Yearly Change'] = r.find_all('td')[3].text
    dict['Net Change'] = r.find_all('td')[4].text
    dict['Density'] = r.find_all('td')[5].text
    
    population_list.append(dict)
```

```python
print(population_list)
```

```python
df = pd.DataFrame(population_list)
df.to_excel('World_population.xlsx', index=False)
```
